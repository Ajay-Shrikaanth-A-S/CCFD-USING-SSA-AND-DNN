```python
# Step 1: Load the Excel file
import pandas as pd

data = pd.read_excel('/content/default of credit card clients.xls')
print("Initial data:")
print(data.head())

# Step 2: Handle missing values (filling with median)
numerical_columns = data.select_dtypes(include=['number']).columns
data[numerical_columns] = data[numerical_columns].fillna(data[numerical_columns].median())
print("After handling missing values:")
print(data.head())

# Identify categorical columns and their cardinality
# Exclude 'Unnamed: 0' as it's likely an index and not a true categorical feature
categorical_columns = data.select_dtypes(include=['object']).columns
categorical_columns = categorical_columns[categorical_columns != 'Unnamed: 0']  # Exclude 'Unnamed: 0'
print("Categorical columns and their cardinality:")
for col in categorical_columns:
    print(f"{col}: {data[col].nunique()} unique values")

# Step 3: Encode categorical variables (Efficiently)
from sklearn.preprocessing import OrdinalEncoder

# Separate high and low cardinality columns
high_cardinality_cols = [col for col in categorical_columns if data[col].nunique() > 10]
low_cardinality_cols = [col for col in categorical_columns if data[col].nunique() <= 10]

# Investigate and address mixed data types in high_cardinality_cols
for col in high_cardinality_cols:
    print(f"Data types in {col}: {data[col].apply(type).unique()}") # Check data types in each column
    # Apply necessary conversions or cleaning here if mixed types are found.
    # For example, if you find integers that should be strings, convert the entire column to strings:
    if int in data[col].apply(type).unique():
        data[col] = data[col].astype(str)

# Ordinal encode high-cardinality columns (after resolving mixed types)
encoder = OrdinalEncoder()
data[high_cardinality_cols] = encoder.fit_transform(data[high_cardinality_cols])

# One-hot encode low-cardinality columns
data = pd.get_dummies(data, columns=low_cardinality_cols, drop_first=True)
print("After encoding categorical variables:")
print(data.head())

# Step 4: Normalize numerical features (Min-Max Scaling)
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
numerical_features = data.select_dtypes(include=['int64', 'float64']).columns
data[numerical_features] = scaler.fit_transform(data[numerical_features])
print("Preprocessed data:")
print(data.head())
```
